# Task 3: Image to Pencil Sketch with Python:

1. Read the image in RBG format using cv2 library
2. Convert it to a grayscale image (black and white)
3. Convert it to a inverted image 
4. Convert it to a blured image 
5. Mix the grayscale image with the inverted blured image

Done! converted to a pencil sketch!


## Importing libraries

```python
import cv2
from google.colab.patches import cv2_imshow
```

## Reading the image



```python
image = cv2.imread("/content/Skyline.png")
cv2_imshow(image)
```

## Converting a BGR image into RGB image

```python
rgbimg = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2_imshow(rgbimg)
```

## Converting the RGB image into gray scale image

```python
grayimg = cv2.cvtColor(rgbimg, cv2.COLOR_RGB2GRAY)
cv2_imshow(grayimg)
```

## Inverting the grey scale image

```python
invertimg = cv2.bitwise_not(grayimg)
cv2_imshow(invertimg)
```

## Blurring the image

```python
blurimg= cv2.GaussianBlur(invertimg, (15, 15), sigmaX=0, sigmaY=0)
cv2_imshow(blurimg)
```

## Pencil Sketching

```python
pencil_sketch = cv2.divide(grayimg, 255 - blurimg, scale=256)
```

```python
cv2_imshow(pencil_sketch)
```
